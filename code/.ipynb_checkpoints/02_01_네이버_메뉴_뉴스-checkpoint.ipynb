## 네이버 메뉴 추출
- 방법 : 크롬 개발자 도구 이용
- 전체 소스 코드에서 위치 찾아 (화면에서 해당 컨텐츠의 단축 메뉴 검사 이용)
- 동일 태그가 여려 개인 경우 class 속성 이용해서 추출

```python
from urllib.request import urlopen
import bs4
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
url = 'https://www.naver.com'

#url에 접속

html = urlopen(url)
type(html)
```

```python
##html.read()
```

```python
# 파서에 등록해 obj로 변환
bs_obj = bs4.BeautifulSoup(html, 'html.parser')
bs_obj
```

```python
# 첫번쨰 메뉴 블럭 파싱 (ul class : list_nav type_fix)
ul = bs_obj.find('ul', {'class':'list_nav type_fix'}) ## 네이버 웹페이지에 존재하는 class 중 하나를 추출하는 과저어
print(ul)

## 위에서 추출한 ul 객체는 bs4.element.Tag임
## 집합적 자료의 속성을 갖고 있기 때문에 반복요소로 사용 가능

```

```python
for li in ul:
    print(li)
```

```python
#findAll 함수 사용해서 ul안의 모든 li 추출
lis = ul.findAll('li')
lis
```

```python
# 각 li 태그 내부의 a태그 접군 후 text 출력
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text, end=', ')
```

```python
ul1 = bs_obj.find('ul',{'class':'list_nav NM_FAVORITE_LIST'})
print
```

```python
lis = ul1.findAll('li')
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text, end=', ')
```

## 네이버 뉴스 크롤링

```python
url = 'https://news.naver.com/'
html = urlopen(url) ## RemoteDisconnected: Remote end closed connection without response
# 프로그램을 통해 자동화된 bot으로 생각하고 연결이 끊어짐 (네이버 정책으로 크롤링 불간능)
```

- 위 코드 처럼 자동화 봇으로 보고 연결을 끊는 경우에는 데이터 전송시 추가되는 header를 구성해 bot이 아님을 증명해야 함
    - header에는 브라우저를 통한 접속이라는 정보를 포함시킨다
    - 크롬 개발자도구 => NetWork => news.naver.com 선택한 후 오른쪽에 새로나타나는 창에서 header 선택
    - requests headers 목록에서 user-agent 확인
- https://developers.whatismybrowser.com/useragents/explore/software_name/chrome/


```python
# header 전송은 보통 requests 객체를 사용함
import requests
headers = {'User-Agent':
           "Mozilla/5.0(Window NT 10.0;Win64;x64) AppleWebKit/537.36 (KHTML,likeGecko) Chrome/87.0.4280.88 Safari/537.36"}
## 딕셔너리 형태로 헤더 구성

url = 'https://news.naver.com/'
res = requests.get(url, headers=headers)
html = res.text
html
```

```python
# bs4 객체 변환
bs_obj = bs4.BeautifulSoup(html, 'html.parser') #bs4에서 사용하기 위한 객체
bs_obj
```





- 전체를 표현하는 코드를 찾고 진행

```python
# 헤드라인 뉴스의 구성 ul, class : hadline__article_list

news_list = bs_obj.find('ul',{'class':'hdline_article_list'})
news_list
```

```python
# 모든 li 추출
lis = news_list.findAll('li')

# 뉴스 제목 출력
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text)
```

```python
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text.strip())
```

```python
# 관련기사 링크 list에 저장
link_list=[]
for li in lis:
    a_tag = li.find('a')
    link_list.append('https://news.naver.com/' + a_tag['href']) ## 동일한 주소가 반복되기 때문에 news.naver.com 생략됨
print(link_list)
' '.join(link_list)
```

## 연습문제
네이버 뉴스 메인페이지의 5개 섹션의 헤드라안이 기사의 타이틀과 링크주소 수집 df생성

```python
import requests

headers ={"User-Agent" : 
          "Mozilla/5.0(Window NT 10.0;Win64;x64) AppleWebKit/537.36 (KHTML,likeGecko) Chrome/87.0.4280.88 Safari/537.36"}

url = "https://news.naver.com/"
res = requests.get(url,headers=headers)
bs_obj = bs4.BeautifulSoup(html,"html.parser")
```

```python
# 수집 데이터 저장할 fltmxm
title_list = []
link_list = []
section_list = []
sections = ["헤드라인","정치", "경제", "사회", "생활/문화", "세계", "IT/과학"]
```

```python
# 7개 섹션 기사를 모두 담고 있는 영역 : div id = main_content 추출
main_cont = bs_obj.find(id='main_content')
uls = main_cont.findAll('ul')
len(uls)

# main_cont > 섹션변 ul 태그 > 각 ul마다 5개의 li 태그 > li태그의 a를 추출
```

```python
for k in range(len(uls)): # 각 ul의 index값 : k변수에 저장 (0-6)
    lis = uls[k].findAll('li') # 한개의 ul태그 안에 모든 li태그를 추출
    for li in lis: # 한개의 li를 li변수에 대입
        a_tag = li.find('a')
        title_list.append(a_tag.text.strip())
        if(k<1):
            link_list.append('https://news.naver.com' + a_tag['href'])
        else:
            link_list.append(a_tag['href'])
            
        section_list.append(sections[k])
    
```

```python
import pandas as pd
```

```python
col_dict = {'section':section_list,'new_title':title_list, 'link_url':link_list}
```

```python
news_df = pd.DataFrame(col_dict)
news_df
```
