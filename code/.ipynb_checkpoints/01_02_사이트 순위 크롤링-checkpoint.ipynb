## 이용 빈도가 높은 웹사이트 순위를 제공하는 사이트 크롤링

- 크롤링 사이트
    - https://www.alexa.com/topsites/countries/KR

```python
import requests
from bs4 import BeautifulSoup
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
url = 'https://www.alexa.com/topsites/countries/KR'

html_website_ranking = requests.get(url).text

# html_website_ranking[:2000] ## 정상적으로 넘어왔는지 확인

soup_website_ranking = BeautifulSoup(html_website_ranking,'html.parser') #파서기를 lxml도 사용가능 (lxml는 범용으로 사용가능함)
```

## 개발자도구 사용법
- 검사 기능을 통해 해당 위치 확인



- 코드에 마아스에 위치시키면 해당 화면 확인


```python
# p태그의 요소 안에서 a태그의 요소를 찾음
website_ranking = soup_website_ranking.select('p a')
website_ranking[0:6] # 첫번째 a태그에 순위와 상관없는 다른 자료가 들어가 있음
```

```python
type(website_ranking[0])

type(website_ranking[0].text)
print(website_ranking[0].text) # 태그 내부의 text를 저장하고 있는 속성

type(website_ranking[0].get_text())
print(website_ranking[0].get_text())
```

```python
## 아래와 동일한 표현
lis = []

for elem in website_ranking:
    lis.append(elem.get_text())

list
```

```python
# 모든 랭킹을 가져와서 리스트에 담기
website_ranking_domain = [website_ranking_element.get_text() for website_ranking_element in website_ranking]
## get_text 태그안에 있는 text를 가져오기 위해서 사용
website_ranking_domain[0:6]
len(website_ranking_domain)
```

```python
# 순위와 상관없는 첫번쨰 요소 제거
#website_ranking_domain.pop(0)
website_ranking_domain[0:6]
```

```python
len(website_ranking_domain)
```

```python
# 전체 data 출력 후 확인 
print("[Top Site in South Korea]")
for k in range(len(website_ranking_domain)):
    print("{0} : {1}".format(k+1, website_ranking_domain[k]))
```

```python
import pandas as pd

website_ranking_dict = {'website': website_ranking_domain}

df = pd.DataFrame(website_ranking_dict, columns=['website'], index=range(1, len(website_ranking_domain)+1))
df
```
