## 네이버 메뉴 추출
- 방법 : 크롬 개발자 도구 이용
- 전체 소스 코드에서 위치 찾아 (화면에서 해당 컨텐츠의 단축 메뉴 검사 이용)
- 동일 태그가 여려 개인 경우 class 속성 이용해서 추출

```python
from urllib.request import urlopen
import bs4
```

```python
url = 'https://www.naver.com'

#url에 접속

html = urlopen(url)
type(html)
```

```python
##html.read()
```

```python
# 파서에 등록해 obj로 변환
bs_obj = bs4.BeautifulSoup(html, 'html.parser')
bs_obj
```

```python
# 첫번쨰 메뉴 블럭 파싱 (ul class : list_nav type_fix)
ul = bs_obj.find('ul', {'class':'list_nav type_fix'}) ## 네이버 웹페이지에 존재하는 class 중 하나를 추출하는 과저어
print(ul)

## 위에서 추출한 ul 객체는 bs4.element.Tag임
## 집합적 자료의 속성을 갖고 있기 때문에 반복요소로 사용 가능

```

```python
for li in ul:
    print(li)
```

```python
#findAll 함수 사용해서 ul안의 모든 li 추출
lis = ul.findAll('li')
lis
```

```python
# 각 li 태그 내부의 a태그 접군 후 text 출력
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text, end=', ')
```

## 네이버 뉴스 크롤링

```python
url = 'https://news.naver.com/'
html = urlopen(url) ## RemoteDisconnected: Remote end closed connection without response
# 프로그램을 통해 자동화된 bot으로 생각하고 연결이 끊어짐 (네이버 정책으로 크롤링 불간능)
```

- 위 코드 처럼 자동화 봇으로 보고 연결을 끊는 경우에는 데이터 전송시 추가되는 header를 구성해 bot이 아님을 증명해야 함
    - header에는 브라우저를 통한 접속이라는 정보를 포함시킨다
    - 크롬 개발자도구 => NetWork => news.naver.com 선택한 후 오른쪽에 새로나타나는 창에서 header 선택
    - requests headers 목록에서 user-agent 확인
- https://developers.whatismybrowser.com/useragents/explore/software_name/chrome/


```python
# header 전송은 보통 requests 객체를 사용함
import requests
headers = {'User-Agent':
           "Mozilla/5.0(Window NT 10.0;Win64;x64) AppleWebKit/537.36 (KHTML,likeGecko) Chrome/87.0.4280.88 Safari/537.36"}
## 딕셔너리 형태로 헤더 구성

url = 'https://news.naver.com/'
res = requests.get(url, headers=headers)
html = res.text
html
```

```python
# bs4 객체 변환
bs_obj = bs4.BeautifulSoup(html, 'html.parser') #bs4에서 사용하기 위한 객체
bs_obj
```





- 전체를 표현하는 코드를 찾고 진행

```python
# 헤드라인 뉴스의 구성 ul, class : hadline__article_list

news_list = bs_obj.find('ul',{'class':'hdline_article_list'})
news_list
```

```python
# 모든 li 추출
lis = news_list.findAll('li')

# 뉴스 제목 출력
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text)
```

```python
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text.strip())
```

```python
# 관련기사 링크 list에 저장
link_list=[]
for li in lis:
    a_tag = li.find('a')
    link_list.append('https://news.naver.com/' + a_tag['href']) ## 동일한 주소가 반복되기 때문에 news.naver.com 생략됨
print(link_list)
' '.join(link_list)
```

```python
# header 전송은 보통 requests 객체를 사용함
import requests
headers = {'User-Agent':
           "Mozilla/5.0(Window NT 10.0;Win64;x64) AppleWebKit/537.36 (KHTML,likeGecko) Chrome/87.0.4280.88 Safari/537.36"}
## 딕셔너리 형태로 헤더 구성

url = 'https://news.naver.com/'
res = requests.get(url, headers=headers)
html = res.text
html
```

```python
bs_obj = bs4.BeautifulSoup(html, 'html.parser') #bs4에서 사용하기 위한 객체
bs_obj
```

```python
news_list = bs_obj.find('ul',{'class':'hdline_article_list'})
news_list
```

```python
news_list.text
```

```python
# 모든 li 추출
lis = news_list.findAll('li')

# 뉴스 제목 출력
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text)
```

```python
title =[]
for li in lis:
    a_tag = li.find('a')
    title.append(a_tag.text.strip())
    print(a_tag.text.strip())
```
